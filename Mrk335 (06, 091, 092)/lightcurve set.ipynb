```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("0.3_10_tb200(00).csv")  

dt=200
counts = data['RATE']
N=len(counts)
pnum = np.arange(len(counts))
t = [i*dt for i in pnum]

plt.figure(figsize=(10,8))
plt.plot(t,counts,'b')
plt.xlabel("TIME(s)",fontsize=20)
plt.ylabel("RATE(coount/s)",fontsize=20)
plt.title("Lightcurve 2000",fontsize=23)
plt.tick_params(labelsize=15)
plt.show()
```
