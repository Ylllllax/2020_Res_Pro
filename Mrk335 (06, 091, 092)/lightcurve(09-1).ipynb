```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("0.3_10_tb200(091).csv")  

dt=200
counts = data['RATE']
N=len(counts)
pnum = np.arange(len(counts))
t = [i*dt for i in pnum]

plt.figure(figsize=(10,8))
plt.plot(t,counts,'b')
plt.xlabel("TIME(s)")
plt.ylabel("RATE(coount/s)")
plt.title("lightcurve-0.3-10-09(1)")
plt.show()    
```
