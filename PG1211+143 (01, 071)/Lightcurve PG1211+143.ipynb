```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("rate_pg1211_01_031_tb50.csv")  

dt=50
counts = data['RATE']
N=len(counts)
pnum = np.arange(len(counts))
t = [i*dt for i in pnum]

plt.figure(figsize=(10,8))
plt.plot(t,counts,'b')
plt.xlabel("TIME(s)",fontsize=15)
plt.ylabel("RATE(coount/s)",fontsize=15)
plt.title("lightcurve-2001-soft",fontsize=20)
plt.show()   
```
