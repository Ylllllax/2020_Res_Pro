```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.fftpack import fft,ifft
import pandas as pd
from scipy.optimize import minimize
```

```python
data02 = pd.read_csv("energy_2002.csv")  
x02 = data02['x']
y02 = data02['y']
xerr02 = data02['xerr']
yerr02 = data02['yerr']

plt.figure(figsize=(10,8))
plt.loglog()
plt.errorbar(x02, y02, yerr=yerr02, xerr=xerr02, fmt=',r')
plt.xlabel("Energy (keV)",fontsize=15)
plt.ylabel(r"$keV^2 (Photons{\ }cm^{-2}{\ } s^{-1}{\ } keV^{-1})$",fontsize=15)
plt.tick_params(labelsize=13)
plt.show()
```
